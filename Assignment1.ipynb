```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-10, 11, 1)
y = np.arange(-10, 11, 1)
xygrid = np.column_stack([[i, j] for i in x for j in y])

plt.figure(figsize=(10, 10), facecolor="w")
plt.scatter(xygrid[0], xygrid[1], s=36, edgecolor="none")
# Set axis limits
plt.grid(True)
plt.axis("equal")
plt.xlim(-20,20)
plt.ylim(-20,20)
plt.title("Original grid in x-y space")
```

```python
x = np.arange(-10, 11, 1)
y = np.arange(-10, 11, 1)
xygrid = np.column_stack([[i, j] for i in x for j in y])

for i in range(len(xygrid[0])):
    xygrid[0][i] += xygrid[1][i]
        
plt.figure(figsize=(10, 10), facecolor="w")
plt.scatter(xygrid[0], xygrid[1], s=36, edgecolor="none")
# Set axis limits
plt.grid(True)
plt.axis("equal")
plt.xlim(-20,20)
plt.ylim(-20,20)
plt.title("Original grid in x-y space")
```

```python
x = np.arange(-10, 11, 1)
y = np.arange(-10, 11, 1)
xygrid = np.column_stack([[i, j] for i in x for j in y])
xygrid[0] = xygrid[0]+xygrid[1]

plt.figure(figsize=(10, 10), facecolor="w")
plt.scatter(xygrid[0], xygrid[1], s=36, edgecolor="none")
# Set axis limits
plt.grid(True)
plt.axis("equal")
plt.xlim(-20,20)
plt.ylim(-20,20)
plt.title("Original grid in x-y space")
```
